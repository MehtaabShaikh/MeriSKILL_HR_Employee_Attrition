# Data Cleaning and Preprocessing

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.drop(["EmployeeCount","Over18","MaritalStatus"],axis=1)
```

```python
df=df.rename(columns={"JobSatisfaction":"Job_Satisfaction"})
```

```python
## Adding new column "Age Group" to know different age groups

bins = [0,20,40,60, float("inf")] # Define the age bins
labels = ['under 20', '21-40','41-59','above 60']  # labels for the age groups

df["Age Group"]=pd.cut(df["Age"], bins=bins, labels=labels, right=False)
```

```python
df[["Age","Age Group"]]
```

```python
aa=sns.countplot(x="Gender",hue="Attrition",data=df)
for bars in aa.containers:
    aa.bar_label(bars)
```

```python
sns.countplot(x="Department",hue="Attrition",data=df)
```

```python
sns.barplot(x="Age Group",y="MonthlyIncome",hue="Attrition", data=df)
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,12))

plt.suptitle("Distribution of various Factors",fontsize=15, color ="green")
sns.histplot(df["Age Group"],ax=ax[0,0])
ax[0][0].set_title("Distribution of Age Group")
sns.histplot(df["PercentSalaryHike"],ax=ax[0,1])
ax[0][1].set_title("Distribution of PercentSalaryHike")
sns.histplot(df["YearsAtCompany"],ax=ax[1,0])
ax[1][0].set_title("Distribution of YearsAtCompany")
sns.histplot(df["YearsSinceLastPromotion"],ax=ax[1,1])
ax[1][1].set_title("Distribution of YearsSinceLastPromotion")

plt.tight_layout()
plt.show()
```

```python
satisfaction_counts=df.groupby("JobRole")["Job_Satisfaction"].value_counts().unstack()
satisfaction_counts.plot(kind="bar", figsize=(14,8))
plt.title("JobRole compared to Job_Satisfaction")

plt.show()
```

```python
df.groupby("EducationField")["Attrition"].value_counts().unstack()
```

```python
ab=df
result = df.groupby(['EducationField'])['Attrition'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.lineplot(x=result.index, y=result.values)
plt.title("EducationField vs Attrition")
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(20,15))
sns.boxplot(x="Job_Satisfaction",y="MonthlyIncome",data=df,hue="JobLevel",palette="Set1")
plt.legend(loc="best")
plt.title("Monthly income on different job level")
plt.show()
```

```python
fig,axes=plt.subplots(figsize=(16,8))
sns.violinplot(x="Attrition",y="YearsAtCompany",hue="Gender",data=df,inner="stick",palette="Set1")
plt.legend(loc="best")
plt.show()
```

```python
non_numeric_columns=df.select_dtypes(include=["object"]).columns
print(non_numeric_columns)
```

```python
df["Attrition"]=df["Attrition"].map({"Yes":1,"No":0})

numeric_df=df.select_dtypes(include=["number"])
fig,ax=plt.subplots(figsize=(14,14))
sns.heatmap(numeric_df.corr(),annot=True,linewidths=4, fmt='.1f',ax=ax,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
